```python
schools_df = pd.read_csv(csv_schools)
students_df = pd.read_csv(csv_students)
```

```python
# Combine the data into a single dataset.
school_data_complete = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
# Calculate the Totals (Schools and Students)
school_count = len(school_data_complete["school_name"].unique())
stundent_count = school_data_complete["Student ID"].count()
# Calculate the Total Budget
total_budget = schools_df["budget"].sum()
```

```python
#print(school_count)
#print(stundent_count)
print(total_budget)
```

```python
# Calculate the Average Scores
average_maths_score = school_data_complete["maths_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()
```

```python
#print(average_maths_score)
#print(average_reading_score)
```

```python
# Calculate the Percentage Pass Rates
passing_maths_count = school_data_complete[(school_data_complete["maths_score"] >= 50)].count()["student_name"]
passing_maths_percentage = passing_maths_count / stundent_count * 100
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 50)].count()["student_name"]
passing_reading_percentage = passing_maths_count / stundent_count * 100
passing_maths_reading_count = school_data_complete[(school_data_complete["maths_score"] >= 50) 
    & (school_data_complete["reading_score"] >= 50)].count()["student_name"]
overall_passing_rate = passing_maths_reading_count / stundent_count * 100 
```

```python
#print(passing_maths_count)
#print(passing_maths_percentage)
#print(passing_reading_count)
#print(passing_reading_percentage)
#print(passing_maths_reading_count)
#print(overall_passing_rate)
```

```python
# Convert to DataFrame
area_summary = pd.DataFrame(
          [{"Total Schools": school_count, 
          "Total Students": stundent_count, 
          "Total Budget": total_budget,
          "Average Math Score": average_maths_score, 
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_maths_percentage,
          "% Passing Reading": passing_reading_percentage,
          "% Overall Passing": overall_passing_rate}])

# Formatting
area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
area_summary
```

```python
#  Calculate the Percentage Pass Rates
per_school_passing_maths = school_data_complete[(school_data_complete["maths_score"] >= 50)]
per_school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 50)]
per_school_passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 50)
                                               & (school_data_complete["maths_score"] >= 50)]

per_school_passing_maths_count = per_school_passing_maths.groupby(["school_name"]).count()["student_name"]
per_school_passing_reading_count = per_school_passing_reading.groupby(["school_name"]).count()["student_name"]
per_school_passing_math_and_reading_count = per_school_passing_math_and_reading.groupby(["school_name"]).count()["student_name"]

# Calculate the average test scores.
per_school_maths = school_data_complete.groupby(["school_name"]).mean(numeric_only=True)["maths_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean(numeric_only=True)["reading_score"]

# Calculate the percentage 
per_school_passing_maths_percentage = per_school_passing_maths_count / per_school_counts * 100
per_school_passing_reading_percentage = per_school_passing_reading_count / per_school_counts * 100
per_school_overall_passing_percentage = per_school_passing_math_and_reading_count / per_school_counts * 100
```
